### Exercice - Traçage d'une courbe

#### Le programme en PYTHON ci dessous permet le traçage de la courbe associée à la fonction $f$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 30)
y = x**2+3*x+5
plt.plot(x, y)
plt.xlabel("abscisses")
plt.ylabel("ordonnees")

plt.show()
```

#### 1. Quelle est l'expression algébrique de la fonction $f$?

$f(x) = x^2+3x+5$
